```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/home/ykalathiya/pycaret_assignment/Student_performance_data _.csv')
data.head()
```

```python
from pycaret.classification import ClassificationExperiment
s = ClassificationExperiment()
```

```python
s.setup(data=data,
        target='GradeClass',
        session_id=123,
        normalize=True,
        transformation=True)
```

```python
best_model = s.compare_models()
```

```python
s.evaluate_model(best_model)
```

```python
# Save the best model for future use
s.save_model(best_model, 'best_student_performance_model')
```

```python
# Predict on the same data (or you can use new/unseen data)
predictions = s.predict_model(best_model)
predictions.head()
```
